# HIGH FREQUENCY WORDS USING NLTK
## Required Packages

```python
import nltk
import pandas as pd
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
import spacy
```

## Corpus

```python

with open('Datasets/headlines_corpus.txt', 'r', encoding='utf-8') as f:
    for line in f:
        print(line)
```

## Data Preprocessing
### Cleaning

```python
import re
line = re.sub(r'[^\w\s]', '',line).lower() # Block-not this characters: words + spaces-Block
print(line)
```

### Tokenization

```python
nltk.download("punkt")
nltk.download("wordnet")
txt_words = nltk.word_tokenize(line)
txt_words
```

### Stopwords

```python
# from nltk.corpus import stopwords
nltk.download("stopwords")
stop_words = stopwords.words('english')
```

```python
# tokenized and clean str
# remove stop words

corpus_final = []
```

```python
for i in txt_words:
    if i not in stop_words:
        corpus_final.append(i)
        corpus_final.append(" ")
```

```python
"".join(corpus_final)
```

### Lemmatization

```python
lemmatizer  = WordNetLemmatizer()
```

```python
for word in corpus_final:
    print(lemmatizer.lemmatize(word))
```

## Frequency Words
### Word Counter

```python
from collections import Counter

word_counter = Counter(corpus_final)
```

### Most Common Words

```python
most_common_words = word_counter.most_common()[:200]
```

```python
print(most_common_words)
```

### Plots

```python
import matplotlib.pyplot as plt
```

```python
sorter_word_counts = sorted(list(word_counter.values()), reverse = True)

plt.loglog(sorter_word_counts)
plt.ylabel("Freq")
plt.xlabel("Word Rank")
```
