NOI- Nodes of interest

```python
import yfinance as  yf
```

Create ticker object to obtain information from yfinance

```python
var1 = yf.Ticker('AAPL')
```

```python
apple = var1.institutional_holders
```

```python
apple['comp'] = var1.ticker
```

```python
var2 = yf.Ticker('MSFT')
```

```python
msft = var2.institutional_holders
```

```python
msft['comp'] = var2.ticker
```

Concatenate two data frames AAPL & MSFT

```python
tog_ = pd.concat([apple, msft])
```

```python
G = nx.from_pandas_edgelist(tog_, 'Holder', 'comp')
```

```python
nx.draw(G, with_labels=True)
```
